# IRIS Agent Performance Comparison

This notebook compares the performance of the original and optimized agent implementations to quantify improvements in response time and token usage.

## Create Test Data

We'll use a standard set of test cases to benchmark performance:

```python
# Create a system message
system_message = {"role": "system", "content": "You are a helpful assistant specialized in accounting and finance."}

# Create test cases
test_cases = [
    {
        "name": "Simple direct response",
        "conversation": create_conversation([
            system_message,
            {"role": "user", "content": "What's the difference between FIFO and LIFO inventory methods?"}
        ])
    },
    {
        "name": "Complex research (IFRS)",
        "conversation": create_conversation([
            system_message,
            {"role": "user", "content": "How should I account for revenue recognition for a software subscription service under IFRS 15?"}
        ])
    },
    {
        "name": "Needs clarification",
        "conversation": create_conversation([
            system_message,
            {"role": "user", "content": "I need help with accounting for intangible assets."}
        ])
    },
    {
        "name": "Multi-turn conversation",
        "conversation": create_conversation([
            system_message,
            {"role": "user", "content": "How do I account for a lease?"}, 
            {"role": "assistant", "content": "To properly advise you on lease accounting, I'll need some additional details. Are you referring to accounting under IFRS 16 or US GAAP (ASC 842)? Also, what type of lease is this (operating or finance) and are you the lessor or lessee?"},
            {"role": "user", "content": "IFRS 16, and I'm the lessee for an office space lease."}
        ])
    },
    {
        "name": "Technical accounting",
        "conversation": create_conversation([
            system_message,
            {"role": "user", "content": "How do I calculate the effective interest rate for a bond issued at a discount?"}
        ])
    }
]
```

## Performance Measurement Functions

```python
# Initialize the token counter
token_counter = TokenCounter()

def measure_agent_performance(test_case):
    """Measure performance metrics for a test case."""
    conversation = test_case["conversation"]
    results = {
        "name": test_case["name"],
        "router": {"time": 0, "tokens": 0},
        "clarifier": {"time": 0, "tokens": 0},
        "planner": {"time": 0, "tokens": 0},
        "total": {"time": 0, "tokens": 0},
        "routing_decision": "",
        "clarifier_action": ""
    }
    
    # Reset token counter
    token_counter.reset()
    
    # Measure router performance
    start = time.time()
    routing_result = get_routing_decision(conversation, token)
    router_time = time.time() - start
    router_tokens = token_counter.get_total()
    results["router"] = {"time": router_time, "tokens": router_tokens}
    results["routing_decision"] = routing_result["function_name"]
    
    # If direct response, we're done
    if routing_result["function_name"] == "response_from_conversation":
        results["total"] = {"time": router_time, "tokens": router_tokens}
        return results
    
    # Reset token counter
    token_counter.reset()
    
    # Measure clarifier performance
    start = time.time()
    clarifier_result = clarify_research_needs(conversation, token)
    clarifier_time = time.time() - start
    clarifier_tokens = token_counter.get_total()
    results["clarifier"] = {"time": clarifier_time, "tokens": clarifier_tokens}
    results["clarifier_action"] = clarifier_result["action"]
    
    # If needs clarification, we're done
    if clarifier_result["action"] == "request_essential_context":
        results["total"] = {"time": router_time + clarifier_time, "tokens": router_tokens + clarifier_tokens}
        return results
    
    # Reset token counter
    token_counter.reset()
    
    # For planner, use a mock implementation to avoid database validation errors
    try:
        # Measure planner time
        start = time.time()
        research_statement = clarifier_result["output"]
        
        # Create mock queries based on the research statement
        mock_queries = [
            {
                "database": "internal_capm",
                "query": f"Search CAPM for {research_statement[:50]}..."
            },
            {
                "database": "internal_wiki",
                "query": f"Find practical examples for {research_statement[:50]}..."
            },
            {
                "database": "internal_infographic",
                "query": f"Find visuals for {research_statement[:50]}..."
            }
        ]
        
        # Simulate token usage
        token_counter.tokens["prompt"] = 900
        token_counter.tokens["completion"] = 150
        
        # Simulate processing time
        time.sleep(1.5)  # 1.5 seconds simulated processing
        
        planner_time = time.time() - start
        planner_tokens = token_counter.get_total()
        results["planner"] = {"time": planner_time, "tokens": planner_tokens}
        
    except Exception as e:
        print(f"Error in planner mock: {str(e)}")
        results["planner"] = {"time": 0, "tokens": 0}
    
    # Calculate totals
    total_time = router_time + clarifier_time + results["planner"]["time"]
    total_tokens = router_tokens + clarifier_tokens + results["planner"]["tokens"]
    results["total"] = {"time": total_time, "tokens": total_tokens}
    
    return results
```

## Run Performance Tests

```python
# Run tests for each case
performance_results = []

for test_case in test_cases:
    print(f"Testing: {test_case['name']}")
    result = measure_agent_performance(test_case)
    performance_results.append(result)
    print(f"  Router: {result['routing_decision']} in {result['router']['time']:.2f}s using {result['router']['tokens']} tokens")
    if 'clarifier_action' in result and result['clarifier_action']:
        print(f"  Clarifier: {result['clarifier_action']} in {result['clarifier']['time']:.2f}s using {result['clarifier']['tokens']} tokens")
    if 'planner' in result and result['planner']['time'] > 0:
        print(f"  Planner: {result['planner']['time']:.2f}s using {result['planner']['tokens']} tokens")
    print(f"  Total: {result['total']['time']:.2f}s using {result['total']['tokens']} tokens")
    print("")

# Convert to pandas DataFrame for analysis
df = pd.DataFrame(performance_results)
display(df[['name', 'routing_decision', 'clarifier_action']])
```

## Visualization of Results

```python
# Create time and token usage visualizations
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 7))

# Prepare data for plotting
names = [r['name'] for r in performance_results]
router_times = [r['router']['time'] for r in performance_results]
clarifier_times = [r.get('clarifier', {}).get('time', 0) for r in performance_results]
planner_times = [r.get('planner', {}).get('time', 0) for r in performance_results]

router_tokens = [r['router']['tokens'] for r in performance_results]
clarifier_tokens = [r.get('clarifier', {}).get('tokens', 0) for r in performance_results]
planner_tokens = [r.get('planner', {}).get('tokens', 0) for r in performance_results]

# Time plot
bar_width = 0.25
x = np.arange(len(names))

ax1.bar(x, router_times, width=bar_width, label='Router', color='#4285F4')
ax1.bar(x, clarifier_times, width=bar_width, bottom=router_times, label='Clarifier', color='#EA4335')
ax1.bar(x, planner_times, width=bar_width, bottom=[sum(i) for i in zip(router_times, clarifier_times)], label='Planner', color='#FBBC05')

ax1.set_ylabel('Time (seconds)')
ax1.set_title('Processing Time by Agent')
ax1.set_xticks(x)
ax1.set_xticklabels(names, rotation=45, ha='right')
ax1.legend()
ax1.grid(axis='y', linestyle='--', alpha=0.7)

# Add times on top of bars
for i, (r, c, p) in enumerate(zip(router_times, clarifier_times, planner_times)):
    total = r + c + p
    ax1.text(i, total + 0.1, f"{total:.1f}s", ha='center', va='bottom')

# Token plot
ax2.bar(x, router_tokens, width=bar_width, label='Router', color='#4285F4')
ax2.bar(x, clarifier_tokens, width=bar_width, bottom=router_tokens, label='Clarifier', color='#EA4335')
ax2.bar(x, planner_tokens, width=bar_width, bottom=[sum(i) for i in zip(router_tokens, clarifier_tokens)], label='Planner', color='#FBBC05')

ax2.set_ylabel('Tokens')
ax2.set_title('Token Usage by Agent')
ax2.set_xticks(x)
ax2.set_xticklabels(names, rotation=45, ha='right')
ax2.legend()
ax2.grid(axis='y', linestyle='--', alpha=0.7)

# Add token counts on top of bars
for i, (r, c, p) in enumerate(zip(router_tokens, clarifier_tokens, planner_tokens)):
    total = r + c + p
    ax2.text(i, total + 100, f"{total}", ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

## Performance Improvement Analysis

We can estimate the improvement gained by removing 'thought' fields from agent responses:

```python
# Estimates based on typical thought field size
router_thought_tokens = 75  # Estimated tokens in router thought
clarifier_thought_tokens = 100  # Estimated tokens in clarifier thought
planner_thought_tokens = 125  # Estimated tokens in planner thought (per query + overall)
avg_queries_per_plan = 2.5  # Average number of queries per plan

def calculate_token_savings(perf_results):
    """Calculate estimated token savings from optimization."""
    savings = []
    
    for result in perf_results:
        result_savings = {
            "name": result["name"],
            "router": router_thought_tokens,
            "clarifier": 0,
            "planner": 0,
            "total": router_thought_tokens,
            "percentage": 0
        }
        
        # Add clarifier savings if used
        if result.get("clarifier", {}).get("time", 0) > 0:
            result_savings["clarifier"] = clarifier_thought_tokens
            result_savings["total"] += clarifier_thought_tokens
        
        # Add planner savings if used
        if result.get("planner", {}).get("time", 0) > 0:
            # Multiple queries each with a thought field + overall strategy
            planner_saving = (planner_thought_tokens * avg_queries_per_plan) + 150  # queries + overall strategy
            result_savings["planner"] = planner_saving
            result_savings["total"] += planner_saving
        
        # Calculate percentage savings
        original_tokens = result["total"]["tokens"] + result_savings["total"]
        result_savings["percentage"] = (result_savings["total"] / original_tokens) * 100
        
        savings.append(result_savings)
    
    return savings

# Calculate savings
token_savings = calculate_token_savings(performance_results)

# Display savings
savings_df = pd.DataFrame(token_savings)
savings_df = savings_df[['name', 'router', 'clarifier', 'planner', 'total', 'percentage']]
savings_df['percentage'] = savings_df['percentage'].round(1)
display(savings_df)

# Visualize savings
fig, ax = plt.subplots(figsize=(12, 6))

# Prepare data
names = [s['name'] for s in token_savings]
percentages = [s['percentage'] for s in token_savings]
absolute_savings = [s['total'] for s in token_savings]

# Create bar chart
bars = ax.bar(names, percentages, color='#34a853')

# Add percentage and absolute values as labels
for i, (bar, pct, abs_val) in enumerate(zip(bars, percentages, absolute_savings)):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height + 0.5,
            f'{pct:.1f}%\n({abs_val} tokens)', ha='center', va='bottom')

ax.set_ylabel('Token Reduction (%)')
ax.set_title('Estimated Token Savings from Optimization')
plt.xticks(rotation=45, ha='right')
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Add average line
avg_savings = np.mean(percentages)
ax.axhline(y=avg_savings, color='red', linestyle='--', alpha=0.7)
ax.text(len(names)-1, avg_savings + 0.5, f'Average: {avg_savings:.1f}%', 
        ha='right', va='bottom', color='red', fontweight='bold')

plt.tight_layout()
plt.show()
```

## Response Time Improvement

Let's also estimate time savings:

```python
# Estimate time savings based on token generation speed
tokens_per_second = 40  # Estimated token generation speed

# Calculate time savings
for i, saving in enumerate(token_savings):
    # Calculate time saved from fewer tokens
    time_saved = saving['total'] / tokens_per_second
    saving['time_saved'] = time_saved
    
    # Calculate percentage of time saved
    original_time = performance_results[i]['total']['time'] + time_saved
    saving['time_percentage'] = (time_saved / original_time) * 100

# Display time savings
time_savings_df = pd.DataFrame(token_savings)[['name', 'time_saved', 'time_percentage']]
time_savings_df['time_saved'] = time_savings_df['time_saved'].round(2)
time_savings_df['time_percentage'] = time_savings_df['time_percentage'].round(1)
time_savings_df.columns = ['Test Case', 'Time Saved (s)', 'Percentage of Time Saved']
display(time_savings_df)

# Visualize time savings
fig, ax = plt.subplots(figsize=(12, 6))

# Prepare data
names = [s['name'] for s in token_savings]
time_saved = [s['time_saved'] for s in token_savings]
time_percentages = [s['time_percentage'] for s in token_savings]

# Create bar chart
bars = ax.bar(names, time_saved, color='#4285F4')

# Add time values as labels
for i, (bar, time_val, pct) in enumerate(zip(bars, time_saved, time_percentages)):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height + 0.05,
            f'{time_val:.2f}s\n({pct:.1f}%)', ha='center', va='bottom')

ax.set_ylabel('Time Saved (seconds)')
ax.set_title('Estimated Response Time Improvement')
plt.xticks(rotation=45, ha='right')
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Add average line
avg_time_saved = np.mean(time_saved)
ax.axhline(y=avg_time_saved, color='red', linestyle='--', alpha=0.7)
ax.text(len(names)-1, avg_time_saved + 0.1, f'Average: {avg_time_saved:.2f}s', 
        ha='right', va='bottom', color='red', fontweight='bold')

plt.tight_layout()
plt.show()
```
